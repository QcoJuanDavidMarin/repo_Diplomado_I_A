## <span style="color:green">**4. Realice un gráfico de disperción de las variables "pm2.5" y "TEMP" con título, labels apropiados y color rojo para los datos. Describa: ¿Qué se puede inferir de dicho gráfico?**</span>


```python
fig, ax = plt.subplots(figsize=(15,5))
x = db['TEMP']
y = db['pm2.5']
plt.scatter(x,y, c = 'red')
plt.title('"pm2.5 Vs TEMP"')
plt.ylabel('pm2.5')
plt.xlabel('TEMP')
plt.grid(c = 'gray')
```

Cuando la temperatura es baja, entre $-10°c$ y $10°c$, es más probable que haya valores altos de $pm2.5$, por lo tanto, se encontraran días altamente contaminados


## <span style="color:green">**5. Calcule la probabilidad de encontrar en la Base de datos el valor de pm2.5 encima de 120**</span>

```python
# total de registros para la variable pm2.5
print('La variable pm2.5 tiene',db['pm2.5'].shape[0],'registros en total y','\n',
      db[db['pm2.5']>120].shape[0],'registros por encima de 120')

print('La probabilidad de encontrar valores de concentración por encima de 120 para la variable pm2.5 es de','\n',
 ((db[db['pm2.5']>120].shape[0])/(db['pm2.5'].shape[0]))*100,'%')
```

## <span style="color:green">**6. Agrupe por año el valor de la variable "pm2.5", calcule el promedio y ordene la base de datos de manera ascendente por el promedio.**</span>




```python
db_sort = (db.groupby('year')[['pm2.5']]
          .mean()
          .sort_values('pm2.5', ascending = True))
display(db_sort)
```
